```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
```

```python
df=pd.read_csv("auto-mpg.csv")
```

```python
df["horsepower"] = pd.to_numeric(df["horsepower"], errors="coerce")
#converting ? to Nan
```

```python
df=df.dropna()
```

```python
df=df.drop(columns="car name")
```

```python
df
```

```python
y=df["mpg"]
x=df[["horsepower","weight"]]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
poly = PolynomialFeatures(degree=2, include_bias=False)
hp_poly_train = poly.fit_transform(x_train[["horsepower"]])
hp_poly_test = poly.transform(x_test[["horsepower"]])
```

```python
x_train_final = np.hstack([hp_poly_train, x_train[["weight"]].values])
x_test_final = np.hstack([hp_poly_test, x_test[["weight"]].values])
```

```python
model = LinearRegression()
model.fit(x_train_final, y_train)

y_pred = model.predict(x_test_final)
```

```python
print(r2_score(y_test, y_pred))
print(mean_squared_error(y_test, y_pred))
```

```python
user_input = pd.DataFrame({
    "horsepower": [145],
    "weight": [3600]
})

# Polynomial transform on horsepower
hp_poly = poly.transform(user_input[["horsepower"]])

# Combine features (SAME order as training)
user_final = np.hstack([
    hp_poly,
    user_input[["weight"]].values
])
predicted_mpg = model.predict(user_final)

print("Predicted MPG:", predicted_mpg[0])
```
